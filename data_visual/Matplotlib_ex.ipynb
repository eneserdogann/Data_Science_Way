```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
#FUNCTIONAL
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.plot(y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r') # 'r' kırmızı renk verir.

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  # birinci değer soldan uzaklık, ikinci yerden uzaklık, 3 yükseklik 4 genişlik.
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
# fig yazıp çalıştırdığımzda sorunsuz çalışır çünkü fig içinde birden çok tablo(axes) barındırabilir
# fakat axes array olarak çalışır ve her node da bir tane tablo bulunur.
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

plt.tight_layout()

axes[0].plot(x,y)
axes[0].set_title('FIRST')

axes[1].plot(y,x)
axes[1].set_title('SECOND')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(3,3))  # figsize tablonun enini ve boyunu ayarlar.

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2, figsize = (4,3))
fig.tight_layout()

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x,'r')
axes[1].set_title('Second Plot')
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
fig.tight_layout()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='X square')
axes.plot(x,x**3,label='Y cube')

axes.legend(loc=(0.025,0.85))  # 1-10 a kadar sayılarda kullanılabilir (loc=10)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,linewidth=3,alpha=0.5,linestyle='--',marker='o',markersize=20,markerfacecolor='red',markeredgewidth='4',
         markeredgecolor='black')  
# linewidth: çigi kalınlığı (lw) .... alpha: transparencys
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,lw=3,ls='--')

axes.set_xlim([0,1])   # X koordinatında 0,1 arasındaki yakınlaştırılmış
axes.set_ylim([0,2])   # Y koordinatında 0,2 arasında yakınlaştırılmış
```
