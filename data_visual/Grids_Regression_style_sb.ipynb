```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
iris['species'].unique()
```

```python
ir = sns.PairGrid(iris)   #Pairplot Ä± Ã¶zelleÅŸtirebildik
ir.map(plt.scatter)
```

```python
ir.map_diag(sns.histplot)
ir.map_upper(plt.scatter)
ir.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tp = sns.FacetGrid(row='smoker',col='time',data=tips)
tp.map(sns.histplot,'total_bill')
```

```python
tp.map(plt.scatter,'total_bill','tip')
```

# Regression

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='time',palette='coolwarm',aspect=0.6,height=8)
```
