```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plots

```python
sns.displot(tips['total_bill'],kde='False',bins=30)  #kde(kernel density) x eksenindeki grafik -- bins çubuk sayısı
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,hue='sex')  #kind='hex' ile hexagonal tablo veya 'kde' ile coğrafik tablo
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill',data=tips,estimator=np.std)
```
